```python
#importing pandas

import pandas
```

```python
# parameters 

Revenue_Growth_Factor = 0.18
EBITDA_Margin = 0.45
Taxes = 0.21
Capital_Expenditures = 0.03
Working_Capital_Increase = 0.06
EBITDA_Exit_Multiple = 32.70
Discount_Factor = 0.94 
```

```python
#showing the data frame for the dtae

dates = pandas.date_range(start = '2022', end = '2032', freq = 'Y')
dates
```

```python
#revenue grwoth first year

df = pandas.DataFrame(index=dates)
df.loc['2022-12-31',"Revenue"] = 222 * (1 + Revenue_Growth_Factor)
df
```

```python
# revenue grothe throw all years

for i in range (1, len(df)):
    df.iloc[i]["Revenue"] = df.iloc[i -1]["Revenue"] * (1 + Revenue_Growth_Factor)
df
```

```python
#calculating 

df['EBITDA'] = df['Revenue']*EBITDA_Margin
df["Taxes"] = -df ["EBITDA"] * Taxes
df["Capital Expenditure"] = -df ["Revenue"] * Capital_Expenditures
df["Working Capital"] = -df ["Revenue"] * Working_Capital_Increase
df["Cash Flow"] = df.drop(columns="Revenue").sum(axis=1)
df
```

```python
df["Discount Factor"] = Discount_Factor
df["Discount Factor"] = df["Discount Factor"].cumprod()
df
```

```python
df["Present Value Free Cash Flow"] = df["Cash Flow"] * df["Discount Factor"]
df
```

```python
msg = f""""Our projected Present Value of Projected Future Cash Flows are: {df['Present Value Free Cash Flow'].sum().round(0)}.
We are providing revenue guidance in 2032 to be: {round(df.iloc[-1]["Revenue"] * Revenue_Growth_Factor)}.
With an estimated EBITDA around: {round(df.iloc[-1]["Revenue"] * 1.18 * 0.50)}.
The terminal value a this time given a multiple of {EBITDA_Exit_Multiple} and the above EBITDA for 2030, would equal:{round(df.loc["2031", "Revenue"] * 1.18 * EBITDA_Exit_Multiple)}
"""

print(msg)

#The terminal value a this time given a multiple of <the-multiple-from-your-variable> and the above EBITDA for 2030, would equal: <your-terminal-value-here>."""
```
